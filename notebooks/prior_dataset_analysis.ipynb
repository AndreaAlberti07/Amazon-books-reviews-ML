## Exploring the Dataset
In this preliminary examination, our objective is to gain a comprehensive understanding of the dataset through a statistical lens.  
Additionally, we will pay special attention to addressing missing values, as their proper handling significantly influences the construction of a robust machine learning algorithm.

```python
import pandas as pd

# Read the data
#books_data = pd.read_csv('../dataset/books_data.csv')
#books_rating= pd.read_csv('../dataset/Books_rating.csv')

# Read the data - Andrea
books_data = pd.read_csv('../data_gitignore/books_data.csv')
books_rating= pd.read_csv('../data_gitignore/books_rating.csv')
```

## Analisys of book data 

```python
books_data.head()
```

```python
books_data.describe()
```

```python
books_data.info()
```

```python
import numpy as np

na_cols=books_data.columns[books_data.isna().any()].tolist()

mv=pd.DataFrame(books_data[na_cols].isna().sum(), columns=['Number_missing'])
mv['Percentage_missing %']=np.round(100*mv['Number_missing']/len(books_data),2)

full_cols=((books_data.isna().any())==False).tolist()
print(f'The columns not showing missing values are: {books_data.columns[full_cols]}')

mv
```
